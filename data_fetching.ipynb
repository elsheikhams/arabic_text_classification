```python
import pandas as pd
import requests
```

```python
df = pd.read_csv('dialect_dataset.csv')
api_endpoint =  'https://recruitment.aimtechnologies.co/ai-tasks'
my_json = df['id'].apply(str).tolist()
iterations_range = list(range(0,len(my_json),1000))
iterations_range.append(len(my_json))
df_text = pd.DataFrame()
for n in range(0,len(iterations_range) - 1):
    r = requests.post(api_endpoint, json = my_json[iterations_range[n]:iterations_range[n + 1]])
    df_text = pd.concat([df_text, pd.DataFrame({'text':r.json().values()})], ignore_index=True, sort=False)
    display(df_text.tail())
    
```

```python
# saving to csv
df_text.to_csv('retreived_text_data.csv')
```
